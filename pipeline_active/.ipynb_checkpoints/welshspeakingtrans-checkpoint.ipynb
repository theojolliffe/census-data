```python
for area in areas:
    if area['parents'][0]['name']=='Wales':
        x = area['data']['welsh']
        print(x)
        for y in x.keys():
            if y=='value':
                print("| ", y)
                for z in x[y].keys():
                    print("| | ", z)
                    print(x[y][z])
                    x[y][z]['NoWelsh'] = x[y][z]['AllOver3'] - x[y][z]['SpeaksWelsh']
                    x[y][z]['all'] = x[y][z]['AllOver3']
                    del x[y][z]['AllOver3']
                    for w in x[y][z].keys():
                        print("| | | ", w)
                        print("| | | | - ", x[y][z][w])
```
