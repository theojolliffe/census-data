```python
# import some packages and instantiate our graph object
import censusdata
import pandas as pd
import networkx as nx

G = nx.Graph()
```

```python
# Create df from censusdata API
df = censusdata.download('acs5', 
           2015,
           censusdata.censusgeo([('state', '36'), ('county', '*')]),
           ['B08301_001E', 'B08301_010E'])
column_names = ['total_transpo', 'total_public_transpo']
df.columns = column_names
df['percent_public_transpo'] = df.apply(
   lambda row: row['total_public_transpo']/row['total_transpo'], 
   axis = 1)

new_indices = []
county_names = []
for index in df.index.tolist():
        new_index = index.geo[0][1] + index.geo[1][1]
        new_indices.append(new_index)
        county_name = index.name.split(',')[0]
        county_names.append(county_name)
df.index = new_indices
df['county_name'] = county_names

df
```

```python
# Create a function that takes in a dataframe of counties and iterates through it to return the desired list
def process_counties(df):
    final_list = []
    for index, row in df.iterrows():
        node = (str(index), 
          {'percent_public_transpo': row['percent_public_transpo'],
           'county_name': row['county_name']})
        final_list.append(node)
    return final_list
```

```python
# add these nodes to our graph
county_FIPS = process_counties(df)
G.add_nodes_from(county_FIPS)
```

```python
# To add in the adjacency edges we’ll need to read in the CSV adjacency file
adj_df = pd.read_csv('/Users/theojolliffe/Downloads/county_adjacency2010.csv')
```

```python
# trim it down to just New York State
county_FIPS_list = [int(i) for i in list(df.index)]
adj_df = adj_df[adj_df['fipscounty'].isin(county_FIPS_list)]
adj_df = adj_df[adj_df['fipsneighbor'].isin(county_FIPS_list)]
adj_df = adj_df[['fipscounty', 'fipsneighbor']]
```

```python
# define a function to turn the dataframe into a list of tuples while removing any mirrored or self-joined edges
def create_tuples(df):
    all_tuples = list(df.itertuples(index=False, name=None))
    final_tuples = []
    for tuple in all_tuples:
        if (tuple[1],tuple[0]) not in final_tuples:
            if tuple[1] != tuple[0]:
                final_tuples.append((str(tuple[0]),str(tuple[1])))
                
    return final_tuples
```

```python
# pass the adjacency dataframe to it and add the results to our graph
fips_tuples = create_tuples(adj_df)
G.add_edges_from(fips_tuples)
```

```python
# define a quick function to draw graphs
import matplotlib.pyplot as plt
def draw_graph(G):
    plt.close()
    nx.draw(G, with_labels=True, font_weight='bold')
    plt.show()
    return

draw_graph(G)
```
